```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("cleaned_bc_data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()

```

```python
numerical_cols = ["Latitude", "Longitude", "Bedrooms", "Bathrooms", "Square Footage", "Price"]

for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
categorical_cols = ["City", "Property Type"]

for col in categorical_cols:
    plt.figure(figsize=(12,4))
    df[col].value_counts().head(20).plot(kind="bar")
    plt.title(f"Top {col} Categories")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(14,5))
df["City"].value_counts().head(20).plot(kind="bar")
plt.title("Top 20 Cities by Listing Count")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[["Latitude", "Longitude", "Bedrooms", "Bathrooms", "Square Footage", "Price"]].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
